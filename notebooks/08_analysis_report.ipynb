# CELL 6 - Evaluation Metrics 

Three metrics were used:


## 1. BLEU Score

Full Form:

Bilingual Evaluation Understudy


Purpose:

Measures word overlap between generated text and reference response.


Higher BLEU means closer wording.



---


## 2. Slot Accuracy


Purpose:

Checks whether important slot values are preserved.


Example:


Expected:

ORD841324


Generated:

ORD841324


Result:

Correct



---


## 3. Exact Match


Purpose:

Checks whether generated response exactly matches the expected sentence.


It is a strict metric because small wording changes cause failure.

# CELL 9 - Production Recommendation

For real-world Conversational AI:


Decoder based systems are suitable when:

- Domain is fixed
- Responses are highly controlled
- Low latency is required
- Compute resources are limited



LLM based systems are suitable when:

- Inputs vary frequently
- Better generalization is needed
- Dynamic slot values must be preserved



Recommended Hybrid Architecture:


Intent + Slots

# CELL 10 - # Final Conclusion


The experiment demonstrates the trade-off between a custom Transformer Decoder and a pretrained LLM.


The Decoder Transformer:

- Achieved higher BLEU score (0.597)
- Learned response templates effectively
- Generated fluent sentences


However:

- It struggled with unseen slot copying
- Slot accuracy was only 0.267



The FLAN-T5 LLM:

- Achieved higher slot accuracy (0.867)
- Preserved important values better
- Required no model training


However:

- Generated different wording
- Had lower BLEU score



Final Observation:

BLEU alone is not sufficient for evaluating task-oriented dialogue generation.

For slot-conditioned NLG systems, factual correctness and slot preservation are more important.


A hybrid LLM + validation approach provides the best practical solution.
